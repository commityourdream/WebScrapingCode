```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
import csv
session = requests.Session()

hdr = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/46.0.2490.80 Safari/537.36',
                       'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8','Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
                       'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
                       'Accept-Encoding': 'none',
                       'Accept-Language': 'en-US,en;q=0.8',
                       'Connection': 'keep-alive'}
#output_file = open("gizbot_reviews_28_feb.txt",'w+')
reviews_links = open("gizbot_urls.txt", 'r')
next = reviews_links.readline()
FinalList=[]
count = 0
reviews_links_list = []
#print >> output_file, "review-id" + "~" +"review-title" + "~" + "review_link" + "~" + "pros" + "~" + "cons" + "~" + "review_text" + "~" + "author" + "~" + "review_date" + "~" + "star-rating"
while next != "": 
    reviews_links_list.append(next)
    next = reviews_links.readline()
    FinalList=[]
for x in reviews_links_list:
    page_getter = str(x).strip()
    print page_getter
    req = session.get(page_getter, headers = hdr)
    soup = BeautifulSoup(req.text,"lxml")
    review_id = re.findall(r'\d+', page_getter)
    #print review_id
    rtitle = soup.find("h1", {"class":"articleheading"}).get_text().replace('\n', "").replace('\t', "").replace('\r', "").strip().encode('utf-8')
    #print rtitle
#     try:
#         pros=[]
#         Parent = soup.find_all("div",{"class": "pros-block"})
#         for product in Parent:
#             for items in product.find_all("ul",{"class": "pros-cons-lists"}):
#                 for tags in items.find_all('li'):
#                     pros.append(tags.text)
#     except:
#         pros = "NA"
#     try:
#         cons=[]
#         Parent = soup.find_all("div",{"class": "cons-block"})
#         for product in Parent:
#             for items in product.find_all("ul",{"class": "pros-cons-lists"}):
#                 for tags in items.find_all('li'):
#                     cons.append(tags.text)
#     except:
#         cons = "NA"
    try:
        rdate = soup.find("div",{"class":"time-date"}).get_text().replace('\n', "").replace('\t', "").replace('\r', "").strip().encode('utf-8')
        print rdate
    except:
        rdate = None
    try:
        review_text=''
        Parent= soup.find("div",{"class":"content-article"})
        for tags in Parent.find_all("p"):
            para=tags.text
            review_text=review_text+" "+para
        print review_text
    except:
         review_text =None
    try:
        reviewer=soup.find("div", {"class":"dateTime"}).get_text()
        #print author
    except:
        reviewer =None
    try:
        overall_rating=soup.find("span", {"itemprop":"ratingValue"}).get_text().replace('\n', "").replace('\t', "").replace('\r', "").strip().encode('utf-8')
        #print overall_rating
       
    except:
        overall_rating=None
    source = {'review-id': review_id[-1], 
              'review-title':rtitle,
              'review_link':page_getter,
#               'pros':pros,
#               'cons':cons,
              'review-date':rdate,
              'review-text':review_text,
              'reviewer':reviewer,'overall_rating':overall_rating}
    FinalList.append(source)
    
```

```python
import pandas as pd
import csv
test_df = pd.DataFrame(FinalList)
print(test_df.info())
test_df
```
