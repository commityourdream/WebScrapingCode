```python
URL=[]
for page in range(1,50):
    page_url = base_url+str(page)
    #print(page_url)
    r=requests.get(page_url)
    soup1 = BeautifulSoup(r.text,"html5lib")
    for link in soup1.findAll('a', attrs={'href': re.compile("^https://gadgets.ndtv.com/mobiles/reviews/")}):
        URL.append(link['href'])
        
pd.Series(URL).to_csv('gadgetsndtvurl.csv')
   
```
