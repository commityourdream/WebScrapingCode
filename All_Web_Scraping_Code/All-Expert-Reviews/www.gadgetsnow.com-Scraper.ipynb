```python
session = requests.Session()
hdr = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/46.0.2490.80 Safari/537.36',
                       'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8','Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
                       'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
                       'Accept-Encoding': 'none',
                       'Accept-Language': 'en-US,en;q=0.8',
                       'Connection': 'keep-alive'}
reviews_links = open("gadgetsnow-urls.txt", 'r')
next = reviews_links.readline()
count = 0
reviews_links_list = []
while next != "": 
    reviews_links_list.append(next)
    next = reviews_links.readline()
FinalList=[]
for x in reviews_links_list:
    page_getter = str(x).strip()
    print page_getter
    req = session.get(page_getter, headers = hdr)
    soup = BeautifulSoup(req.text,"lxml")
    review_id = re.findall(r'\d+', page_getter)
    #print review_id
    try:
        rtitle=soup.find("h1", {"class":"heading1"}).get_text()
        #print rtitle
    except:
        rtitle ="NA"
#     try:
#         pros=soup.find("h1", {"class":"pros"}).get_text()
#         print pros
#     except:
#         pros ="NA"
#     try:
#         cons=soup.find("h1", {"class":"pros"}).get_text()
#         print cons
#     except:
#         cons ="NA"
    try:
        rdate=''
        Parent = soup.find("span", {"class":"time_cptn"})
        for tags in Parent.find_all('span'):
            para_text=tags.text
            rdate=rdate+" "+para_text
            print rdate
    except:
        rdate = "NA"
    try:
        review_text=''
        Parent = soup.find_all("div",{"class": "Normal"})
        for tags in Parent:
            para_text=tags.text
            review_text=review_text+" "+para_text
            print review_text
    except:
        review_text="NA"
    try:
        author=soup.find("span",{"class":"auth_detail"}).get_text()
        #print author
    except:
        author="NA"
    try:
        srating=soup.find("span", {"class":"rating_value"}).get_text()
        print srating
    except:
        srating=" NA"
    source = {'review_id': review_id[-1], 
              'review_title':rtitle,
              'review_url':page_getter,
#               'pros':pros,
#               'cons':cons,
              'review_date':rdate,
              'review_text':review_text,
              'reviewer':author,'overall_rating':srating}
    FinalList.append(source)
print("Done")
        
   
        
           
                    
   
        
        
               
   
        
```
