```python
def select_random_proxy(proxy_sf):
   row_num = random.randint(0, len(proxy_sf) - 1)
   proxy_ip = "http://{}:{}@{}:{}".format(proxy_sf.iloc[row_num]['user'],
                                               proxy_sf.iloc[row_num]['pass'],
                                               proxy_sf.iloc[row_num]['ip'],
                                               proxy_sf.iloc[row_num]['port'])
   proxy = {"http": proxy_ip, "https":proxy_ip}
   return proxy
proxies = pd.read_csv('proxy_20_may.csv',sep=':')
```

```python
FinalList=[]


for i in range(len(SourceProID)):
    url=SourceProID['url'].iloc[i].format(1)
    print url
    time.sleep(random.randint(1,5))
    try:
        r = requests.get(url,headers=hdr,proxies=select_random_proxy(proxies))
        print r
        content = r.content.decode(encoding='UTF-8')
        soup = BeautifulSoup(r.content.decode(encoding='UTF-8'), "html5lib")
        review_id = re.findall(r'\d+',url)
        #print review_id
        rtitle = soup.find("h1", {"itemprop":"headline"})
        if rtitle:
            rtitle=rtitle.get_text().replace('\n', "").replace('\t', "").replace('\r', "").strip().encode('utf-8')
            #print rtitle
        else:
            rtitle=None
        rdate = rdate = soup.find("span", {"class":"dtreviewed"})
        if rdate:
            rdate=rdate.get_text().replace('\n', "").replace('\t', "").replace('\r', "").strip().encode('utf-8')
            print rdate
        else:
            rdate=None
        try:
            remove=".embed-container { position: relative; padding-bottom: 56.25%; height: 0; overflow: hidden; max-width: 100%; } .embed-container iframe, .embed-container object, .embed-container embed { position: absolute; top: 0; left: 0; width: 100%; height: 100%; }"
            review_text=''
            Parent= soup.find("div",{"class":"fullstoryCtrl_fulldetails"})
            for tags in Parent.find_all("p"):
                para=tags.text.replace(remove,"")
                review_text=review_text+" "+para
                #print review_text
        except:
            review_text =None
        author=soup.find("span", {"itemprop":"name"})
        if author:
            author=author.get_text()
            print author
        else:
            author=None
        try:
            Parent=soup.find("div", {"class":"avg_rating"})
            #print Parent
            rating= re.findall(r'\d+',str(Parent))
            print rating
        except:
            rating=None
       
        source = {'review_id': review_id[-1], 
                  'review_title':rtitle,
                   'review_url':url,
                   'review_date':rdate,
                   'review_text':review_text,
                   'reviewer':author,'overall_rating':rating}
        FinalList.append(source)
            
            
            
    except ChunkedEncodingError:
        print("ChunkedEncodingError!")
    except ProxyError:
        print("Proxy Error!")
    except ConnectionError:
        print("Connection Error!")
    except Exception as ex:
        print("Error",ex)
print('Done')
```
