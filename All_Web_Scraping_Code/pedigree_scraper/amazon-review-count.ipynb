```python
def select_random_proxy(proxy_sf):
    row_num = random.randint(0, len(proxy_sf) - 1)
    proxy_ip = "http://{}:{}@{}:{}".format(proxy_sf.iloc[row_num]['user'],
                                               proxy_sf.iloc[row_num]['pass'],
                                               proxy_sf.iloc[row_num]['ip'],
                                               proxy_sf.iloc[row_num]['port'])
    proxy = {"http": proxy_ip, "https":proxy_ip}
    return proxy
proxies = pd.read_csv('proxies_15Apr.csv',sep=':')
proxies.head()
```

```python
def select_random_UserAgent(UseragentList):
    row_num = random.randint(0, len(UseragentList) - 1)
    return UseragentList['User-agent'].iloc[row_num]
```

```python
List=pd.read_csv('useragent - Sheet1 (1).csv')
```

```python
headers={'authority':'www.amazon.in',
           'method':'GET',
           'path':'/s?k=smartphones&ref=nb_sb_noss_2',
           'scheme':'https',
           'accept':'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
           'accept-encoding':'gzip, deflate, br',
           'accept-language':'en-GB,en-US;q=0.9,en;q=0.8,pl;q=0.7',
           'cache-control':'max-age=0',
           'referer':'https://www.amazon.in/',
           'upgrade-insecure-requests':'1', 
           'User-Agent':select_random_UserAgent(List)}
```

```python
SourceProID=pd.read_csv('amazon_pedigree_data.csv')

SourceProID['review_count']=None
SourceProID['price']=None
```

```python
SourceProID.head()
```

```python
for i in range(len(SourceProID)):
    url=SourceProID['source_product_url'].iloc[i].format(1)
    print url
    time.sleep(random.randint(1,5))
    headers['User-agent']=select_random_UserAgent(List)
    headers['referer']= SourceProID['source_product_url'].iloc[i].format(1)
    try:
        r = requests.get(url,headers=headers,proxies=select_random_proxy(proxies))
        print r
        content = r.content.decode(encoding='UTF-8')
        soup = BeautifulSoup(r.content.decode(encoding='UTF-8'), "html5lib")
        #print soup
        #parent=soup.find("span",{"id":"priceblock_ourprice"})
        price=soup.find("span",{"id":"priceblock_ourprice"})
        print price
        Rcount=soup.find("span",{"id":"acrCustomerReviewText"})
        print Rcount
        if Rcount:
                SourceProID['review_count'].iloc[i]=Rcount.text
        else:
             SourceProID['review_count'].iloc[i]=None
        if price:
                SourceProID['price'].iloc[i]=price.text
        else:
             SourceProID['price'].iloc[i]=None
        
    except ChunkedEncodingError,ProxyError:
        SourceProID['review_count'].iloc[i]=None
        SourceProID['price'].iloc[i]=None
    except Exception as e:
        print e
print("Done")
```
