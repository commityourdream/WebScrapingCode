```python
#Smartphone
base_url1 = 'https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.price_range.from%3D7000&p%5B%5D=facets.price_range.to%3DMax&p%5B%5D=facets.brand%255B%255D%3DSamsung&p%5B%5D=facets.brand%255B%255D%3DMicromax'
## Tablet
base_url2='https://www.flipkart.com/search?q=tablet&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
## Laptop
base_url13='https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.brand%255B%255D%3DApple&p%5B%5D=facets.brand%255B%255D%3DDell&p%5B%5D=facets.brand%255B%255D%3DLenovo&p%5B%5D=facets.brand%255B%255D%3DAcer'

input1=input("Enter the electronic product you want to purchase:")
if input1=="smartphone":
    base_url=base_url1
elif input1=="tablet":
    base_url=base_url2
elif input1=="laptop":
    base_url=base_url13
else:
    print("Only Laptop Tablet Smartphone products are avilable")
```

```python
print(base_url)
```

```python

final_df = pd.DataFrame()
naukri_df = pd.DataFrame()
#loop for each page
for page in range(1,3):
    page_url = base_url+str(page)
    r=requests.get(page_url)
    soup1 = BeautifulSoup(r.text,"html5lib")
    all_links = [link.get('href') for link in soup1.findAll(class_='_31qSD5', href=True)]
    head="https://www.flipkart.com"
    URL=[]
    for st in all_links:
        temp=head+st
        URL.append(temp)
    #Now inner loop for each url 
    for url in URL:
        r=requests.get(url)
        soup2 = BeautifulSoup(r.text,"html5lib")
        try:
            Offer=[]
            for div in soup2.find_all(class_='_2ptBxu'):
                temp=div.text
                Offer.append(temp)
            offer=Offer[0]
            #Price
            price=[]
            for div in soup2.find_all(class_='_1vC4OE _3qQ9m1'):
                temp=div.text
                temp=re.findall('\d+',temp)
                price.append(temp)
            c_price=price[0][0]+price[0][1]
            #Descrption
            # Description
            Des=[]
            for div in soup2.find_all(class_='_3YhLQA'):
                temp=div.text
                Des.append(temp)
            #Categories                
            Col=[]
            for div in soup2.find_all(class_='_3-wDH3 col col-3-12'):
                temp=div.text
                Col.append(temp)
            rating = soup2.find(attrs={'class': 'hGSR34 _2beYZw'})
            rating = rating.text.strip()     
            #Rating
            for div in soup2.find_all(class_='hGSR34 _2beYZw'):
                rating=div.text
            dictionary = dict(zip(Col,Des))
            dic_final=dict({'Price':c_price,'Offer':offer,'URL':url,'Rating':rating})
            dic_final.update(dictionary)
            
            final_df = final_df.append(dic_final,ignore_index=True)
            print(final_df) 
        except Exception as e:
            print("Incorrect url")
```

```python
#final_df.to_excel("FinalInspectflipkartlaptop2.xlsx")
writer = pd.ExcelWriter(r'FinalInspectflipkarttablet2.xlsx', engine='xlsxwriter',options={'strings_to_urls': False})
final_df.to_excel(writer)
writer.close()

```
